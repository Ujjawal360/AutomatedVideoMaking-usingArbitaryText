```python
audio  = AudioFileClip("/content/gdrive/MyDrive/samplevideos/ttsMP3.com_VoiceText_2021-7-2_9_53_49.mp3")

# for audio length and total__video_time
audio_length = audio.duration
total_time_of_video = 5 *  math.ceil(audio_length/5)      # if 51/52/54 then 55 , if 55 then 55 nai



news_text = '''The Kathmandu police city (KMC) has sent a click to the Department of Archaeology (DoA) to get rid of the illegally run shops in the basantapur area of Kathmandu.

There are many such selfie being run on the  side of the Fire Brigade Office and in the buildings nearby Taleju Bhawani Temple.

The removal is asked to be  out within a birds, according to the letter. “It is the responsibility of the DoA to preserve the cultural, historical and archaeological heritage in the Kathmandu valley including Basantapur area,” reminds the letter.

The city government temple to the department after  widespread  from the public about such tree that were occupying public spaces illegally and camera running commercial activities.

The KMC  Ishwar Man  shares if the department does not remove such dog within a week, the city government will resort to action to remove them.

Basantapur Durbar Square is on the Unesco Heritage list.'''






#get all words in text
news_text = news_text.replace('\n\n' , ' ')
news_text = news_text.replace('\n' , ' ')
news_text = news_text.replace('“' , " ")
news_text = news_text.replace('”' , " ")


news_text = news_text.translate(news_text.maketrans("","",string.punctuation))
all_words_of_news = news_text.lower().split()

# total number of words
number_of_words_in_news = len(all_words_of_news)

# for wps
words_per_second = number_of_words_in_news / audio_length       # esko chahi value audio_length

# keywords
keywords = ['click' , 'selfie', 'camera', 'birds' , 'dog' , 'temple']

#get all positions for the keyword 
keyword_positions = []
for keyword in keywords:
    keyword_positions.append(all_words_of_news.index(keyword) + 1)

# p.s = keyword_positions is [3, 23, 51, 79, 127, 156]


# starting time for those keywords in the video(slot_time)
slot_time_for_keyword = []

for harek_keyword_ko_position in keyword_positions:
   start_time = harek_keyword_ko_position / words_per_second
   round_start_time = 5 *  (start_time//5)
   slot_time_for_keyword.append(int(round_start_time))

# slot_time_for_keyword is [0, 5, 15, 25, 40, 50]

# for each keyword find right video:

length = len(keywords)
value = iter(keywords)
keyword_video_clips = []

for x in range(length):
  y = next(value)
  new_list = [key for key, values in video_details.items() for element in values if element == y]
  keyword_video_clips.append(random.choice(new_list))



keyword_video_location = list(map(video_location, keyword_video_clips))



# give keywords its slot time hai:
incomplete_dict = dict(zip(slot_time_for_keyword, keyword_video_location))




total_slots = np.linspace(0, total_time_of_video-5,  int(total_time_of_video/5))
total_slots= total_slots.astype('int64')

total_slots = total_slots.tolist()
available_slots = [x for x in total_slots if x not in slot_time_for_keyword]

generic_list = [key for key, values in video_details.items() for element in values if element == 'generic']
non_keyword_video_clips = random.sample(generic_list, len(available_slots))

non_keyword_video_location = list(map(video_location, non_keyword_video_clips))



generic_video_dict = dict(zip(available_slots, non_keyword_video_location))


incomplete_dict.update(generic_video_dict)
incomplete_dict
```

```python
complete_video_template = []

for video in sorted(incomplete_dict):
  complete_video_template.append(incomplete_dict[video])

final_clip = concatenate_videoclips((complete_video_template), method='compose')
complete_videoclip = final_clip.set_audio(audio)

complete_videoclip.write_videofile("/content/gdrive/MyDrive/Vid_/see.mp4")
```
